### Predictive model to determine which passengers were more likely to survive the sink in GENDER/SEX columns.

```python
titanic["Sex"]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
titanic["Sex"] = label_encoder.fit_transform(titanic["Sex"])
titanic.head()
```

```python
# Create a count plot to visualize survival counts by gender with the same color map
sns.countplot(x=titanic['Sex'], hue=titanic['Survived'], palette='viridis')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Survivals by Gender')
plt.legend(title='Survived')
plt.show()
```

### Here I Gonna Drop the Non-Required Column

```python
titanic.isnull().sum()
```

```python
# dropping non required columns
titanic = titanic.drop(['Age'], axis=1)
```

```python
titanic.head()
```

### Predictive model to determine which passengers were more likely to survive the sink in PCLASS (Passenger Class) columns.

```python
# Count the number of passengers in each class
pclass_counts = titanic['Pclass'].value_counts()

# Create a pie chart
plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Passenger Classes')

plt.show()
```

```python
# Create a count plot to visualize survival counts by passenger class
sns.countplot(x=titanic['Survived'], hue=titanic['Pclass'])

# Add labels and title
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Count of Survivals by Passenger Class')
plt.legend(title='Passenger Class')
plt.show()
```

#### Here I'm doing the model training

```python
# Select features and target variable
a = titanic[['Pclass', 'Sex']]
b = titanic['Survived']
```

```python
# Import necessary libraries for splitting data, building the model, evaluating the model, and handling missing values
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
```

```python
# Split the data
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=0)
```

```python
log = LogisticRegression(random_state=0)
log.fit(a_train, b_train)
```

#### Here I'm doing the model prediction

```python
# Predict and evaluate the model
b_pred = log.predict(a_test)
b_pred
```
